```python
data1=data[['Hotel_Name' , 'Hotel_Address' , 'Tags' , 'Average_Score']]
```

```python
data1['country']=data1.Hotel_Address.apply(lambda x: x.split(' ')[-1] )
```

```python
data1.loc[:, 'country'] = data1['country'].str.replace("Netherlands", "NL")
data1.loc[:, 'country'] = data1['country'].str.replace("Kingdom", "UK")
data1.loc[:, 'country'] = data1['country'].str.replace("France", "FR")
data1.loc[:, 'country'] = data1['country'].str.replace("Spain", "ES")
data1.loc[:, 'country'] = data1['country'].str.replace("Italy", "IT")
data1.loc[:, 'country'] = data1['country'].str.replace("Austria", "AT")
```

```python
data1.head()
```

```python
#remove_puctuation
def impute(column):
    column=column[0]
    if (type(column) != list):
        return "".join(literal_eval(column))
    else:
        return column
```

```python
data1["Tags"]=data[["Tags"]].apply(impute ,axis=1)
data1.head()
```

```python
#lowercase words 
data1['Tags']=data1['Tags'].str.lower()
data1['country']=data1['country'].str.lower()
data1.head()
```

```python
#recommender function 
def recommender(location , description):
    #dividing sentence to words
    description = description.lower()
    word_tokenize(description)
    #apply stopword 
    stop_words=stopwords.words('english')
    #apply lemmatization
    lemm=WordNetLemmatizer()
    #connect packege and filter
    filterd = {word for word in description if not word in stop_words}
    filterd_set = set()
    for fs in filterd :
        filterd_set.add(lemm.lemmatize(fs))
        #create variable that takes location 
    country=data1[data1['country']==location.lower()]
    country=country.set_index(np.arange(country.shape[0]))
    list1=[] ; list2=[] ; cos=[];
    for i in range(country.shape[0]):
        temp_token=word_tokenize(country["Tags"][i])
        temp_set=[word for word in temp_token if not word in stop_words]
        temp2_set = set()
        for s in temp_set:
            temp2_set.add(lemm.lemmatize(s))
        vector=temp2_set.intersection(filterd_set)
        cos.append(len(vector))
        #apply cosine similarity 
    country['similarity']=cos
    country=country.sort_values(by='similarity' , ascending = False)
    country.drop_duplicates(subset='Hotel_Name' , keep='first' , inplace=True)
    country.sort_values('Average_Score' , ascending = False , inplace = True)
    country.reset_index(inplace=True)
    return country[["Hotel_Name","Average_Score","Hotel_Address"]].head(10)
```

```python
recommender('UK' ,'Business trip')
```
